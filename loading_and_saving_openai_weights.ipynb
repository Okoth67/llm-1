```python
#gpt2 model
#GPT Model
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np

# ========== Helper Function ==========
def assign(left, right):
    if left is None:
        raise ValueError("Left tensor is None (e.g., model weight not initialized)")
    if right is None:
        raise ValueError("Right tensor (from weights) is None (e.g., missing in params)")
    if left.shape != right.shape:
        raise ValueError(f"Shape Mismatch. Left: {left.shape}, Right: {right.shape}")
    left.data.copy_(torch.tensor(right))
    return left

# ========== GPT2 Model Definition ==========
class MultiHeadAttention(nn.Module):
    def __init__(self, embed_dim, num_heads, dropout=0.1):
        super().__init__()
        assert embed_dim % num_heads == 0, "Embedding dimension must be divisible by number of heads"
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads

        self.qkv_proj = nn.Linear(embed_dim, 3 * embed_dim)
        self.o_proj = nn.Linear(embed_dim, embed_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        B, T, C = x.size()
        qkv = self.qkv_proj(x).view(B, T, 3, self.num_heads, self.head_dim)
        q, k, v = qkv.unbind(dim=2)
        q, k, v = q.transpose(1, 2), k.transpose(1, 2), v.transpose(1, 2)

        attn_scores = (q @ k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn_scores = attn_scores.masked_fill(torch.tril(torch.ones(T, T, device=x.device)) == 0, float('-inf'))
        attn_weights = F.softmax(attn_scores, dim=-1)
        attn_weights = self.dropout(attn_weights)

        out = attn_weights @ v
        out = out.transpose(1, 2).contiguous().view(B, T, C)
        return self.o_proj(out)

class FeedForward(nn.Module):
    def __init__(self, embed_dim, dropout=0.1):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(embed_dim, 4 * embed_dim),
            nn.GELU(),
            nn.Linear(4 * embed_dim, embed_dim),
            nn.Dropout(dropout)
        )

    def forward(self, x):
        return self.layers(x)

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, dropout=0.1):
        super().__init__()
        self.attn = MultiHeadAttention(embed_dim, num_heads, dropout)
        self.ffn = FeedForward(embed_dim, dropout)
        self.norm1 = nn.LayerNorm(embed_dim)
        self.norm2 = nn.LayerNorm(embed_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        x = x + self.dropout(self.attn(self.norm1(x)))
        x = x + self.dropout(self.ffn(self.norm2(x)))
        return x

class GPTModel(nn.Module):
    def __init__(self, vocab_size, max_seq_len, embed_dim, num_heads, num_layers, dropout=0.1):
        super().__init__()
        self.embed_dim = embed_dim
        self.token_emb = nn.Embedding(vocab_size, embed_dim)
        self.pos_emb = nn.Embedding(max_seq_len, embed_dim)
        self.trf_block = nn.Sequential(*[
            TransformerBlock(embed_dim, num_heads, dropout)
            for _ in range(num_layers)
        ])
        self.dropout = nn.Dropout(dropout)
        self.final_norm = nn.LayerNorm(embed_dim)
        self.lm_head = nn.Linear(embed_dim, vocab_size)

    def forward(self, x):
        B, T = x.size()
        tok_embed = self.token_emb(x)
        pos_ids = torch.arange(T, device=x.device).unsqueeze(0)
        pos_embed = self.pos_emb(pos_ids)
        x = tok_embed + pos_embed
        x = self.trf_block(x)
        x = self.dropout(x)
        x = self.final_norm(x)
        logits = self.lm_head(x)
        return logits
```

```python
#first define a small utility function to check whether right and left tensors have same dimension the copy the model weights
def assign(left, right):
    if left is None:
        raise ValueError("Left tensor is None (e.g., model weight not initialized)")
    if right is None:
        raise ValueError("Right tensor (from weights) is None (e.g., missing in params)")
    if left.shape != right.shape:
        raise ValueError(f"Shape Mismatch. Left: {left.shape}, Right: {right.shape}")

    if isinstance(right, torch.Tensor):
        right_tensor = right.detach().clone()
    else:
        right_tensor = torch.tensor(right)
    left.data.copy_(right_tensor)
    return left
```

```python
def load_weights_into_gpt(gpt, params):
    assign(gpt.pos_emb.weight, params['wpe'])
    assign(gpt.token_emb.weight, params['wte'])

    for b in range(len(params['blocks'])):
        p_block = params['blocks'][b]
        block = gpt.trf_block[b]

        q_w, k_w, v_w = np.split(p_block['attn']['c_attn']['w'], 3, axis=-1)
        q_b, k_b, v_b = np.split(p_block['attn']['c_attn']['b'], 3, axis=-1)

        assign(block.attn.qkv_proj.weight.data[:gpt.embed_dim], torch.tensor(q_w.T))
        assign(block.attn.qkv_proj.weight.data[gpt.embed_dim:2*gpt.embed_dim], torch.tensor(k_w.T))
        assign(block.attn.qkv_proj.weight.data[2*gpt.embed_dim:], torch.tensor(v_w.T))

        assign(block.attn.qkv_proj.bias.data[:gpt.embed_dim], q_b)
        assign(block.attn.qkv_proj.bias.data[gpt.embed_dim:2*gpt.embed_dim], k_b)
        assign(block.attn.qkv_proj.bias.data[2*gpt.embed_dim:], v_b)

        assign(block.attn.o_proj.weight, p_block['attn']['c_proj']['w'].T)
        assign(block.attn.o_proj.bias, p_block['attn']['c_proj']['b'])

        assign(block.ffn.layers[0].weight, p_block['mlp']['c_fc']['w'].T)
        assign(block.ffn.layers[0].bias, p_block['mlp']['c_fc']['b'])
        assign(block.ffn.layers[2].weight, p_block['mlp']['c_proj']['w'].T)
        assign(block.ffn.layers[2].bias, p_block['mlp']['c_proj']['b'])

        assign(block.norm1.weight, p_block['ln_1']['g'])
        assign(block.norm1.bias, p_block['ln_1']['b'])
        assign(block.norm2.weight, p_block['ln_2']['g'])
        assign(block.norm2.bias, p_block['ln_2']['b'])

    # Final norm and output head (check if keys exist)
    if 'ln_f' in params:
        assign(gpt.final_norm.weight, params['ln_f']['g'])
        assign(gpt.final_norm.bias, params['ln_f']['b'])
    else:
        print("Warning: 'ln_f' key missing in params — skipping final norm weights loading.")

    assign(gpt.lm_head.weight, params['wte'])
```

```python
#loading the gpt2 weights ie calling the fucntion
# Define model
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

gpt = GPTModel(
    vocab_size=50257,
    max_seq_len=1024,
    embed_dim=768,
    num_heads=12,
    num_layers=12
)

load_weights_into_gpt(gpt, params)
gpt.to(device)

```
